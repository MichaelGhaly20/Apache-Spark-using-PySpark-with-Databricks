```
data = [
("James", 25),
("Michael", 26) ,
("Robert",30),
("Maria", 28),
("Jen",33)
]
```

```
from pyspark.sql import SparkSession
from pyspark.context import SparkContext

spark = SparkSession.builder.getOrCreate()
df = spark.createDataFrame(data,['name','age'])

df.printSchema()
df.show()
```

### Method 1: Using Lpad

```
from pyspark.sql.functions import lpad, format_string, concat, lit, substring

df_new  = df.withColumn('Age000', lpad('age', 3, '0')).show()
```

### Method 2: Format_string

```
df_new  = df.withColumn('Age000', format_string("%03d",'age')).show()
```

```
df_new  = df.withColumn('Age000', format_string("%s#%03d",'name', 'age')).show()
```

### Method 3: Concat and Substring

```
df_new  = df.withColumn('Age000', concat(lit("00"), 'age')).show()
```

```
df_new  = df.withColumn('Age000', substring('age', -3, 3)).show()
```
