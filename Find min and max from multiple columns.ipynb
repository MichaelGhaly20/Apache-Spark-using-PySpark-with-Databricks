```
list_data=[
    ["mike", 20,33,60,44],
    ["james", 58, 33, 78,83] ,
    ["john", 75,81,56,67],
    ["sam", 100, 100,93,87],
    ["sara", 7,32,44, 18],
    ["kelly", 53,25, 55,41]
]
list_Schema = ["name", "Term_1", "Term_2", "Term_3", "Term_4"]
```

```
df = spark.createDataFrame(list_data, list_Schema)

df.show()
```

### Methods
##### 1. Using Spark SQL function - Greatest() & Least()
##### 2. Using Spark UDF (User Defined Functions)

```
from pyspark.sql.functions import greatest,least

df_new = df.withColumn("Max",greatest("Term_1", "Term_2", "Term_3", "Term_4")) \
         .withColumn("Min", least("Term_1", "Term_2", "Term_3", "Term_4"))

df_new.display()
```

```
@udf
def min_max(type, col:list):
    if type == 'Max':
        return max(col)
    else:
        return min(col)
```

```
from pyspark.sql.functions import array,lit

df_new = df.withColumn("Max",min_max(lit("Max"), array(["Term_1", "Term_2", "Term_3", "Term_4"]))) \
         .withColumn("Min", min_max(lit("Min"), array(["Term_1", "Term_2", "Term_3", "Term_4"])))

df_new.display()
```
